```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 5
variance = 1.2
sigma = math.sqrt(variance)
x = np.linspace(-5, 15, 1000)

plt.figure(figsize=(8, 4))
plt.xticks([5])
plt.yticks([0])
plt.ylim([-1.1, 1.1])
plt.plot(x, 2*stats.norm.pdf(x, mu, sigma), label="Input Distribution")
plt.plot(x, np.tanh(x), label="Activation function")
plt.legend()
plt.show()

plt.figure(figsize=(8, 4))
plt.xticks([5])
plt.yticks([0])
plt.ylim([-1.1, 1.1])
plt.plot(x, 2*stats.norm.pdf(x, mu, sigma), label="Input Distribution")
plt.plot(x, np.tanh(x-5), label="Activation function")
plt.legend()
plt.show()
```
