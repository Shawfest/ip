```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
mu = 5
variance = 1.2
sigma = math.sqrt(variance)
x = np.linspace(-5, 15, 1000)

plt.figure(figsize=(8, 4))
plt.xticks([5])
plt.yticks([0])
plt.ylim([-1.1, 1.1])
plt.plot(x, 2*stats.norm.pdf(x, mu, sigma), label="Input Distribution")
plt.plot(x, np.tanh(x), label="Activation function")
plt.legend()
plt.show()

plt.figure(figsize=(8, 4))
plt.xticks([5])
plt.yticks([0])
plt.ylim([-1.1, 1.1])
plt.plot(x, 2*stats.norm.pdf(x, mu, sigma), label="Input Distribution")
plt.plot(x, np.tanh(x-5), label="Activation function")
plt.legend()
plt.show()
```

# Infomax

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
l = np.linspace(-5, 5, 10000)
x = sigma * np.random.randn(l.shape[0]) + mu
x = np.sort(x)
beta = 0
alpha = 1
eta = 0.001

for i in range(100):
    y = np.tanh((alpha*x + beta))
    Ex = np.mean(x)
    Exx = np.mean(x**2)
    Ey = np.mean(y)
    Eyy = np.mean(y**2)
    Exy = np.mean(x*y)
    alpha = alpha + eta * (1/alpha - 2*Exy)
    beta = beta + eta * (-2*Ey)
    
print(beta)
print(alpha)

plt.figure(figsize=(8,8))
n, bins, patches = plt.hist(x=y, bins='auto', alpha=0.7, rwidth=0.85)
plt.show()
```
